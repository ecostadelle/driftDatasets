```python
def filter(data: pd.DataFrame, column_name: str, alpha: float):
    '''
    Function creates a neu column with filtered data

    Parameters
    ----------
        data = Pandas DataFrame with the data
        column_name
    
    Returns
    -------
        updates the pandas DataFrame 
    '''
    filtered = []
    beta = 1 - alpha
    last_data_value = 0
    for row in data.index:
        data_value = data.x[row]
        new_data_value = last_data_value*beta+data_value*alpha
        last_data_value = new_data_value
        filtered.append(new_data_value)
    data["new_"+column_name] = filtered
    return sns.lineplot(data["new_"+column_name], dashes=False)
```

```python
filter(SEA_training_data, "x", 0.001)
#sns.lineplot(SEA_training_class[500:1000], dashes=False)
```
